```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np
    
# Open and read the file calc-kpoint.dat
kp, ener = np.loadtxt('calc-kpoint.dat', delimiter=' ', unpack=True)

# Create figure object
plt.figure()
# Plot the data, using black color
plt.scatter(kp, ener, s=30, c='blue')
# Plot a dashed line at 14x14x1
plt.axhline(ener[10], c='black', ls='-.', label='12x12x1 k-point convergence', alpha=0.7)
plt.annotate(f'Convergence at {ener[10]} Ry', xy=(140, ener[10]), xytext=(100, ener[10]+0.001))
plt.legend()

# Set the axis limits
plt.xlim(0, 200)
plt.ylim(-23.913, -23.905)
# Add the x and y-axis labels
plt.xlabel('k-points mesh density')
plt.ylabel('Total energy (Ry)')
plt.title('Total energy vs Number of k-points in graphene')
# Save figure to the pdf file
plt.savefig('plot-kpoint.pdf')
# Show figure
plt.show()
```
